```python
def fatorial(n):
  if n == 0:
    return 1
  else:
    fat = 1
    for i in range(1, n + 1):
      fat *= i

    return fat

```

```python
nome = input("Digite seu nome: ")
print(nome)

num1 = float(input("Qual o primeiro número? "))
operacao = input("Qual operação quer realizar: +, -, * || /: ")
num2 = float(input("Qual o segundo número? "))
if operacao == "+":
    print(num1 + num2)
elif operacao == "-":
    print(num1 - num2)
elif operacao == "*":
    print(num1 * num2)
elif operacao == "/":
    print(num1 / num2)
```
